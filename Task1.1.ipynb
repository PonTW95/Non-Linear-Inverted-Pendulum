```
import numpy as np
import autograd.numpy as np
from matplotlib.pyplot import ion, draw, Rectangle, Line2D
import matplotlib.pyplot as plt
from CartPole import *
import random
from scipy import interpolate
```

```
#Task 1.1
##Initialise State vector
#state = [cart location, cart vel, pole angle, pole velocity]
n= 500
def rollout():
    f = CartPole()
    f.getState()
    states=np.zeros([n,4])
    # Simple Oscillation
    # Vary initial state to get different behaviours
    states[0]=f.setState([0,0,np.pi,0])
    for i in range(n):
        states[i]=f.getState()
        #perform equation of motion
        f.performAction()
        #map angle into accepted range
        f.remap_angle()
    t=np.arange(0.0, n*0.2 , 0.2)
    labels = ['Cart Location', 'Cart Velocity', 'Pole Angle', 'Pole Velocity']  
    x_new = np.linspace(0,100,100000)
    # interpolate to smooth curve
    sp = interpolate.make_interp_spline(t, states[:,0])
    y_new_1 = sp(x_new)
    plt.xlabel('time(s)')
    plt.ylabel(labels[0])
    #plt.xlim(0,15)
    #plt.ylim(-10,5)
    plt.plot(x_new, y_new_1,color='royalblue')
    plt.title("State Variation with Time with Complete Rotation")
    plt.show()
    x_new = np.linspace(0,100,100000)
    sp = interpolate.make_interp_spline(t, states[:,1])
    y_new_2 = sp(x_new)
    plt.xlabel('time(s)')
    plt.ylabel(labels[1])
    #plt.xlim(0,10)
    #plt.ylim(-5,5)
    plt.plot(x_new, y_new_2,color='royalblue')
    plt.title("State Variation with Time with Complete Rotation")
    plt.show()
    x_new = np.linspace(0,100,100000)
    sp = interpolate.make_interp_spline(t, states[:,2])
    y_new = sp(x_new)
    plt.xlabel('time(s)')
    plt.ylabel(labels[2])
    #plt.xlim(0,5)
    #plt.ylim(-7,7)
    plt.plot(x_new, y_new,color='royalblue')
    plt.title("State Variation with Time with Complete Rotation")
    plt.show()
    sp = interpolate.make_interp_spline(t, states[:,3])
    y_new_4 = sp(x_new)
    #plt.xlim(0,5)
    plt.plot(x_new, y_new_4,color='royalblue')
    plt.xlabel('time(s)')
    plt.ylabel(labels[3])
    #plt.ylim(-10,20)
    plt.title("State Variation with Time with Complete Rotation")
    plt.show()
    plt.plot(states[:,2],states[:,3])
    plt.title("Phase plot of pole angular velocity against pole angle")
    plt.xlabel("Pole Angle")
    plt.ylabel("Pole Velocity")
    plt.show()
    plt.plot(y_new_1,y_new_2,'--')
    plt.title("Phase plot of Cart Velocity against Cart Location")
    plt.xlabel("Cart Location")
    plt.ylabel("Cart Velocity")
    plt.xlim(-10,0)
    plt.show()
    plt.scatter(states[:,1],states[:,2])
    plt.title("Phase plot of Pole Angle against Cart Velocity")
    plt.xlabel("Cart Velocity")
    plt.ylabel("Pole Angle")
    plt.ylim(-5,5)
    plt.show()
    plt.plot(y_new_2,y_new_4,'--')
    plt.title("Phase plot of Pole Angular Velocity against Cart Velocity")
    plt.xlabel("Cart Velocity")
    plt.ylabel("Pole Angular Velocity")
    plt.ylim(-10,10)
    plt.show()
    plt.plot(y_new_1,y_new,'--')
    plt.title("Phase plot of Pole Angle against Cart Location")
    plt.xlabel("Cart Location")
    plt.ylabel("Pole Angle")
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
    plt.plot(y_new_1,y_new_4,'--')
    plt.title("Phase plot of Pole Angular Velocity against Cart Location")
    plt.xlabel("Cart Location")
    plt.ylabel("Pole Anglular Velocity")
    plt.xlim(-10,10)
    plt.ylim(-15,20)
    plt.show()
```

```
rollout()
```
